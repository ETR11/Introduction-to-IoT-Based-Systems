```python
#Import libraries
import pandas as pd #https://pandas.pydata.org/
import matplotlib.pyplot as plt #https://matplotlib.org/
import matplotlib.dates as mdates #https://matplotlib.org/
```

```python
#March
speed_20180729 = pd.read_csv("march.csv",dayfirst=True,sep=",",
                           header=0,decimal=b".",index_col=0, 
                           parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])
                          
speed_20180729.info()

date = '2022-03-15 Jokioinen Ilmala-observation station'
plt.figure(figsize=(16,6)) 
plt.plot(speed_20180729, color='blue', marker='.',linestyle='-') 
plt.title(date) 
plt.ylabel("Direct solar radiation (W/m2)") 
plt.grid(True) 
plt.show() 
```
