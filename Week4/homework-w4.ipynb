```python
import networkx as nx #https://networkx.github.io/
import matplotlib.pyplot as plt #https://matplotlib.org/
import scipy as sp
import numpy as np
import pandas as pd
```

```python
#Creating a graph
G = nx.Graph()
```

```python
#Adding nodes
G.add_node("A")
#Two or more nodes
G.add_nodes_from(["B", "C", "D", "E"])
```

```python
#Creating edges
G.add_edges_from([("A", "B"), ("A", "C"), ("A", "D"), ("A", "E")])
```

```python
#Plot the graph
nx.draw_networkx(G, node_color='black', font_color='white', edge_color='black', font_weight='bold',  node_size=700)
plt.axis('off')
plt.show()
print("The reflexive system: A + Ab + Ac + Ad + Ae = A(1 + b + c + d + e)")
print("The structure of awareness of the system: (1 + b + c + d + e)")
```

```python
Temperature1 = pd.read_csv("airport_temp.csv",dayfirst=True,sep=",", 
                          header=0,decimal=b".",index_col=0,
                          parse_dates= [[0, 1, 2, 3]],usecols=[0,1,2,3,5])
Temperature1.info()
```

```python
plt.figure(figsize=(16,6)) 
plt.plot(Temperature1['2018-7-20'] > 25,color='black', marker='o',linestyle='') 
plt.title("From node B to node A. Air temperature in July 20, 2018, at Lappeenranta airport station") 
plt.ylabel("Node B sends 1 when air tamperature is > 25") 
plt.grid(True) 
plt.show() 
```
